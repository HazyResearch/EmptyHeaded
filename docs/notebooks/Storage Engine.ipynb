# EmptyHeaded Storage Engine

This tutorial shows how to view the code that is generated from EmptyHeaded's query compiler. This is the code that is calls and is executed in the storage engine.


Make sure you have run the Getting Started Tutorial before executing the following command.

```python
from emptyheaded import *
start()
triangle="""
TriangleAgg(;z) :- Edge(a,b),Edge(b,c),Edge(a,c),z:long<-[COUNT(*)].
"""
eh=os.path.expandvars("$EMPTYHEADED_HOME")
db = Database.from_existing(eh+"/docs/db")
ir = db.eval(triangle)
stop()
```

Lets inspect the intermediate representation (or query plan) that EmptyHeaded uses to generate. This corresponds to the image above. The generated code for this query is in:

$EMPTYHEADED_HOME/docs/db/libs/**folder**/run_0.hpp  

The **folder** is named with the date and time at which you ran the query. So search for the folder that corresponds to when you ran the query.
